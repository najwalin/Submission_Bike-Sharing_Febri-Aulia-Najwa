## Menentukan Pertanyaan Bisnis

**SMART Question** adalah sebuah framework untuk merumuskan pertanyaan secara terstruktur agar memeperoleh informasi yang mendalam.

* **Spesific (Spesifik)**
  * Pertanyaan harus jelas, fokus pada sebuah topik tertentu, dan tidak bermakna ganda. Hindari pertanyaan yang terlalu luas.
    * Salah: Bagaimana cara meningkatkan penjualan?
    * Benar: Faktor apa saja yang memengaruhi penurunan penjualan produk kategori elektronik di wilayah Jakarta selama kuartal terakhir?
* **Measurable (Terukur)**
  * Pertanyaan harus bisa dijawab dengan angka atau matrix yang konkret, harus tahu apa yang akan dihitung.
    * Salah: Apakah pelanggan senang dengan layanan kita?
    * Benar: Berapa skor rata-rata Customer Satisfaction untuk layanan purna jual bulan ini dibandingkan bulan lalu?
* **Action-Oriented (Berorientasi Aksi)**
  * Hasil dari pertanyaan harus bisa memberikan arahan untuk melakukan tindakan nyata. Jika pertanyaan terjawab, stakeholder harus tahu apa langkah selanjutnya.
    * Salah: Mengapa orang suka berbelanja?
    * Benar: Fitur apa pada aplikasi yang paling sering digunakan sebelum pengguna memutuskan untuk melakukan checkout?
* **Relevant (Relevan)**
  * Hasil dari pertanyaan harus sejalan dengan tujuan utama bisnis atau masalah yang sedang dihadapi.
    * Salah: Menanyakan tentang stok gudang saat masalah utamanya adalah efektivitas kampanye media sosial.
    * Benar: Apakah kampanye iklan di Instagram memberikan Return on Ad Spend (ROAS) yang lebih tinggi dibandingkan iklan di TikTok?
* **Time-bound (Terikat Waktu)**
  * Pertanyaan harus ada batasan waktu yang jelas agar analisis memiliki konteks yang tepat.
    * Salah: Berapa banyak pengguna baru kita?
    * Benar: Berapa tingkat pertumbuhan pengguna baru secara bulanan (Month-over-Month) sepanjang tahun 2025?

### Cleaning Data

#### Fixing ... problem

```python
# Menampilkan tipe data dteday menjadi datetime pada data day_df
datetime_colums = ["dteday"]
for column in datetime_colums:
  day_df[column] = pd.to_datetime(day_df[column])

day_df.info()
```

```python
# Menampilkan tipe data dteday menjadi datetime pada data hour_df
datetime_colums = ["dteday"]
for column in datetime_colums:
  hour_df[column] = pd.to_datetime(hour_df[column])

hour_df.info()
```

```python
season_map = {1:'Winter', 2:'Spring', 3:'Summer', 4:'Fall'}
weather_map = {1:'Clear', 2:'Cloudy', 3:'Light Rain', 4:'Heavy Rain'}
weekday_map = {0:'Sunday', 1:'Monday', 2:'Tuesday', 3:'Wednesday', 4:'Thursday', 5:'Friday', 6:'Saturday'}
year_map = {0: '2011', 1:'2012'}

day_df['season'] = day_df['season'].map(season_map)
day_df['weathersit'] = day_df['weathersit'].map(weather_map)
day_df['weekday'] = day_df['weekday'].map(weekday_map)
day_df['yr'] = day_df['yr'].map(year_map)

hour_df['season'] = hour_df['season'].map(season_map)
hour_df['weathersit'] = hour_df['weathersit'].map(weather_map)
hour_df['weekday'] = hour_df['weekday'].map(weekday_map)
hour_df['yr'] = hour_df['yr'].map(year_map)
hour_df['day_type'] = hour_df['workingday'].map({1: 'Weekday', 0:'Weekend'})

day_df.head()
hour_df.head()
```

**Insight:** (Opsional)
- Kolom `dteday` pada dataset `day_df` dan `hour_df` telah berhasil dikonversi dari tipe `object` menjadi **datetime64[ns]**.
- Perubahan ini memungkinkan data tanggal digunakan secara optimal dalam analisis berbasis waktu, seperti pengelompokan berdasarkan hari, bulan, maupun tahun.
- Beberapa variabel numerik telah diubah menjadi bentuk kategorikal yang lebih mudah dipahami, yaitu:
  * season menjadi: Winter, Spring, Summer, Fall
  * weathersit menjadi: Clear, Cloudy, Light Rain, Heavy Rain
  * weekday menjadi: Sunday, Monday, Tuesday, Wednesday, Thursday,
  Friday, Saturday.
  * yr menjadi: 2011, 2012
- Perubahan ini mengubah representasi data dari numerik menjadi lebih kategorikal, sehingga lebih mudah dibaca dan diinterpretasikan dalam analisis.
- Selain itu, proses pemetaan dilakukan secara konsisten tanpa menghasilkan nilai kosong (`NaN`), yang menunjukkan bahwa seluruh kategori berhasil terdefinisi dengan baik.
- Ditambahkan variabel `day_type` pada dataset `hour_df`, yang mengelompokkan hari menjadi:
  * Weekday (hari kerja)
  * Weekend (akhir pekan)
- Variabel ini membantu dalam menganalisis perbedaan pola penyewaan antara weekday dan weekend.
- Tidak terdapat perubahan jumlah baris maupun kolom setelah proses cleaning.
- Seluruh data tetap utuh tanpa kehilangan informasi, sehingga integritas data tetap terjaga.

## Analisis Lanjutan (Opsional)

```python
# Clustering waktu berdasarkan aktivitas harian
def time_category(hr):
  if 5 <= hr < 10:
    return 'Morning'
  elif 10 <= hr < 15:
    return 'Afternoon'
  elif 15 <= hr < 20:
    return 'Evening'
  else:
    return 'Night'

hour_df['time_category'] = hour_df['hr'].apply(time_category)

# Rata-rata penyewaan sepeda berdasarkan kategori waktu
time_cluster = hour_df.groupby('time_category')['cnt'].mean().reset_index()
time_cluster

# Rata-rata penyewaan berdasarkan kategori eaktu dan tipe hari
time_daytype = hour_df.groupby(['time_category', 'day_type'])['cnt'].mean().reset_index()
time_daytype
```

```python
# Visualisasi Clustering
plt.figure(figsize=(8,6))

sns.barplot(
    data=time_cluster,
    x='time_category',
    y='cnt',
    order=['Morning', 'Afternoon', 'Evening', 'Night']
)

plt.title('Rata-Rata Penyewaan Sepeda Berdasarkan Kategori Waktu')
plt.xlabel('Kategori Waktu')
plt.ylabel('Rata-Rata Penyewaan Sepeda')
plt.show()
```
